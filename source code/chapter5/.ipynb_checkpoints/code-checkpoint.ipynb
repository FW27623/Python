## 5.2 简单数据信息的可视化

```python
# 导入库
import matplotlib.pyplot as plt
plt.rc("font",family="SimHei",size="14")  #用于解决中文显示乱码问题
plt.rcParams['axes.unicode_minus'] =False
import pandas as pd
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
# 获取原始数据
raw_data = pd.read_excel('demo.xlsx')
raw_data.head()
```

### 5.2.1 使用条形图和柱形图表达数据差异

1. 柱形图

```python
province_data = raw_data.groupby(['PROVINCE'],as_index=False).sum()
province_bar = province_data.sort_values(['AMOUNT'],ascending=False)
# 柱形图
province_bar.plot(kind='bar', x='PROVINCE',y=['AMOUNT','MONEY'], figsize=(10, 4),title='各省份商品销售对比',fontsize=12)
```

2. 条形图

```python
province_barh = province_data.sort_values(['AMOUNT'],ascending=True)
# 条形图
province_barh.plot(kind='barh', x='PROVINCE',y=['AMOUNT','MONEY'], figsize=(10, 4), logx=True,title='各省份商品销售对比',fontsize=10)
```

### 5.2.2 使用折线图和柱形图展示趋势

1. 折线图

```python
datetime_data = raw_data.groupby(['DATETIME'],as_index=False).sum()
datetime_data.plot(kind='line', x='DATETIME',y=['AMOUNT','MONEY'], figsize=(10, 4),title='按日销售走势')
```

2. 柱形图

```python
# 从时间中提取月份
raw_data['MONTH'] = raw_data['DATETIME'].map(lambda i: i.month)
# 基于月份做汇总
datemonth_data = raw_data.groupby(['MONTH'],as_index=False).sum()
# 柱形图展示时间序列趋势
datemonth_data.plot(kind='bar', x='MONTH',y=['AMOUNT','MONEY'], figsize=(10, 4),title='按月份销售走势')
```

### 5.2.3 使用饼图和面积图展示成分或结构信息

1. 饼图

```python
# 按销售类型汇总
cate_data = raw_data.groupby(['CATE'],as_index=False)['VISITS'].sum()
cate_data = cate_data.sort_values(['VISITS'],ascending=False)
labels = cate_data['CATE']
cate_data.plot(kind='pie', y='VISITS', figsize=(6, 6),title='VISIT在各个CATE中的分布', labeldistance=1.1, autopct="%1.1f%%", shadow=False, startangle=90, pctdistance=0.6,labels=labels,legend=False)
```

2. 面积图

```python
area_data = raw_data.groupby(['MONTH'],as_index=False).sum()
area_data['S1_PEC'] = area_data['STORE1_AMOUNT']/area_data['AMOUNT']
area_data['S2_PEC'] = area_data['STORE2_AMOUNT']/area_data['AMOUNT']
area_data.plot(kind='area', x='MONTH',y=['S1_PEC','S2_PEC'], figsize=(10, 4),alpha=0.5,title="'STORE1_AMOUNT','STORE2_AMOUNT'销售占比趋势")
```

### 5.2.4 使用散点图展示数据间关系

1. 散点图

```python
raw_data.plot(kind='scatter',x='AMOUNT', y='VISITS', figsize=(10, 4),title='AMOUNT和MONEY关系')
```

2. 蜂窝图

```python
raw_data.plot(kind='hexbin', x='AMOUNT',y='MONEY', gridsize=10, figsize=(8, 6),title='销售关系分析')
```

### 5.3.9 使用漏斗图展示不同转化环节的完成情况

漏斗图

```python
# 导入pyecharts库
from pyecharts import options as opts
from pyecharts.charts import Funnel
import pandas as pd
```

```python
# 读取sheet2中的数据
raw_data2 = pd.read_excel('demo.xlsx',sheet_name =1)
raw_data2.head(3)
```

```python
# 转换为漏斗所用数据
data = raw_data2.drop(['DATE','IMPRESSIONS'],axis=1).sum(axis=0)
funnel_data = [[i,j] for i,j in zip(data.index,data.to_list())]
print(funnel_data)
```

```python
# 漏斗可视化图表
funnel=Funnel()
funnel.add("流量转化漏斗", funnel_data)
funnel.set_global_opts(title_opts=opts.TitleOpts(title="广告点击、到达和转化漏斗"))
funnel.render_notebook()
```

### 5.3.10 使用关系图展示不同元素见的关联关系 

关系图

```python
from pyecharts.charts import Graph

# 读取原始数据
raw_data3 = pd.read_excel('demo.xlsx',sheet_name =2)
raw_data3.head(3)
```

```python
# 找到所有的节点，即前项和后项商品集合并添加
node_data = raw_data3['前项商品'].append(raw_data3['后项商品'])
value_count = node_data.value_counts()
nodes = [opts.GraphNode(name=node, symbol_size=int(np.log(value)*10)) for node,value in zip(value_count.index,value_count.values)]
links = [opts.GraphLink(source=i, target=j) for i,j in raw_data3[['前项商品','后项商品']].values]
```

```python
# 画出关系图
graph = Graph()
graph.add("", nodes, links, repulsion=4000)
graph.set_global_opts(title_opts=opts.TitleOpts(title="商品关系"))
graph.render_notebook()
```

### 5.3.11 使用雷达图展示多个元素在不同属性上的差异

雷达图

```python
# 导入库
from pyecharts.charts import Radar
```

```python
# 计算得到雷达图所用数据
radar_gb = raw_data.groupby('CATE',as_index=False).mean()
names = ['AMOUNT', 'MONEY', 'VISITS', 'PAGEVIEWS','STORE1_AMOUNT','STORE2_AMOUNT']
radar_data = radar_gb[names]
max_value,min_value = radar_data.max(axis=0),radar_data.min(axis=0)
cates = [f'CATE_{i}' for i in radar_gb['CATE'].values]
```

```python
# 画雷达图
rader = Radar()
schema = [opts.RadarIndicatorItem(name=i, max_=j*1.05, min_=k*0.95) for i,j,k in zip(names,max_value,min_value)]
rader.add_schema(
    schema=schema
)
colors = ["#ff0000","#005e15","#00736d","#440062","#a84200"]
for ind,value in enumerate(cates):
    data = [radar_data.iloc[ind].round(0).tolist()]
    rader.add(value, data,color=colors[ind])
rader.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
rader.set_global_opts(title_opts=opts.TitleOpts(title='雷达图对比'))
rader.render_notebook()
```
