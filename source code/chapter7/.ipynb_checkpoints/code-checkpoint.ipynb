## 7.1 使用KMeans聚类算法挖掘用户潜在特征

```python
# 导入库
import pandas as pd # panda库
from sklearn.preprocessing import MinMaxScaler # 标准化库
from sklearn.cluster import KMeans  # 导入sklearn聚类模块
from sklearn.metrics import silhouette_score  # 效果评估模块
```

```python
# 读取数据
cluster_data = pd.read_excel('demo.xlsx',sheet_name=0,index_col='USER_ID')
print(cluster_data.head(3))
```

```python
# 数据预处理
scaler = MinMaxScaler() # 标准化
scaled_features = scaler.fit_transform(cluster_data)
print(scaled_features[:2,:])
```

```python
# 聚类
model_kmeans = KMeans(n_clusters=3, random_state=0)  # 建立聚类模型对象
model_kmeans.fit(scaled_features)  # 训练聚类模型
```

```python
# 模型评估
n_samples, n_features = cluster_data.shape
print('samples: %d \t features: %d' % (n_samples, n_features))
silhouette = silhouette_score(scaled_features, model_kmeans.labels_, metric='euclidean')  # 平均轮廓系数
print('silhouette score:',silhouette)
```

```python
# 组合原始数据与标签
kmeans_labels = pd.DataFrame(model_kmeans.labels_,index=cluster_data.index,columns=['labels']) 
kmeans_data = pd.concat((cluster_data,kmeans_labels),axis=1)
print(kmeans_data.head(3))
```

```python
# 基于聚类群组汇总特征
radar_gb = kmeans_data.groupby('labels',as_index=False).mean()
print(radar_gb)
```

```python
# 柱形图展示
radar_gb.plot(kind='bar', x='labels',y=['AVG_ORDERS','AVG_MONEY','IS_ACTIVE','SEX'], figsize=(10, 4),logy=True, title='不同聚类分组结果对比')
```

```python
# 雷达图展示
from pyecharts.charts import Radar
from pyecharts import options as opts
names = kmeans_data.columns[:-1].tolist()
radar_data = radar_gb[names]
max_value,min_value = radar_data.max(axis=0),radar_data.min(axis=0)
cates = [f'CLUSTER_{i}' for i in radar_gb['labels'].values]
rader = Radar()
schema = [opts.RadarIndicatorItem(name=i, max_=j*1.05, min_=k*0.95) for i,j,k in zip(names,max_value,min_value)]
rader.add_schema(schema=schema)  
colors = ["#ff0000","#005e15","#00736d","#440062"] 
for ind,value in enumerate(cates): 
    data = [radar_data.iloc[ind].round(0).tolist()]
    rader.add(value, data,color=colors[ind]) 
rader.set_series_opts(label_opts=opts.LabelOpts(is_show=False)) 
rader.set_global_opts(title_opts=opts.TitleOpts(title='雷达图对比'))  
rader.render_notebook()  
```

## 7.2 使用决策树预测客户是否会产生转化

```python
# 导入库
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split  
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
# 读取原始数据
tree_data = pd.read_excel('demo.xlsx',sheet_name=1,index_col='USER_ID')
x,y = tree_data.iloc[:,:-1],tree_data.iloc[:,-1]
print(tree_data.head(1))
```

```python
# 特征预处理
# 填充缺失值
null_val = x.isnull().any(axis=0)
print(null_val[null_val==True])
x = x.fillna(x.mean()) 
# 拆分训练集和测试集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=0)  # 将数据分为训练集和测试集
```

```python
# 训练模型
model_tree = DecisionTreeClassifier(max_depth=4,class_weight='balanced')
model_tree.fit(x_train, y_train)  # 训练决策树模型
pre_y = model_tree.predict(x_test)  # 使用测试集做模型效果检验
```

```python
# 模型评估
metrics = [accuracy_score,precision_score,recall_score,f1_score]
scores = [i(y_test, pre_y) for i in metrics]
columns = [i.__name__ for i in metrics]
scores_pd = pd.DataFrame(scores,index=columns).T
print(scores_pd)
```

```python
# 展示特征重要性
feature_pd = pd.Series(model_tree.feature_importances_,index=x.columns)
feature_pd.sort_values().plot(kind='barh',figsize=(10, 4), title='特征重要性评估')
```

```python
# 重新训练模型
new_data = pd.read_csv('tree.txt',index_col='USER_ID')# 重新读预测数据
model_tree.fit(x,y)
print(model_tree.predict(new_data))
proba_data = pd.DataFrame(model_tree.predict_proba(new_data),columns=model_tree.classes_)
print(proba_data)
```

## 7.3 使用主成分分析+岭回归预测广告UV量

```python
# 导入库
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.decomposition import PCA
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder,MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer,mean_absolute_error,mean_squared_error,median_absolute_error
```

```python
# 读取数据
ad_data = pd.read_excel('demo.xlsx',sheet_name=2)
print(ad_data.head(3))
```

```python
# 数据预处理
# 先转数值型，再做onehotencode
cols = ['TYPE','SIZE','PROMOTION']
cate_model = Pipeline([('ode', OrdinalEncoder()), ('ohe', OneHotEncoder(categories='auto',sparse=False))])
cate_data = cate_model.fit_transform(ad_data[cols])

# 数值型标准化
num_model = MinMaxScaler()
num_data = num_model.fit_transform(ad_data[['MONEY']])
# 合并为x
x = np.hstack((cate_data,num_data))
# 拆分y
y = ad_data['UV']
```

```python
# 模型训练
pipe_model = Pipeline([('pca', PCA(n_components=0.85)), ('ridge', Ridge())])
pipe_model.fit(x,y)
```

```python
# 模型评估
metrics = [mean_absolute_error,mean_squared_error,median_absolute_error]
columns = [i.__name__ for i in metrics]
scores = [cross_val_score(pipe_model, x, y, cv=3,scoring=make_scorer(i),n_jobs=-1) for i in metrics]
reg_pd = pd.DataFrame(scores,index=columns).T.round(1)
print(reg_pd)
```

```python
# 柱形图对比结果
reg_pd.plot(kind='bar',figsize=(10, 4), logy=True,title='不同交叉检验结果对比')
```

```python
# 预测新数据
data_new = pd.read_csv('ad.txt')
# 数据预处理
cate_data_new = cate_model.transform(data_new[cols])
num_data_new = num_model.transform(data_new[['MONEY']])
# 合并
x_new = np.hstack((cate_data_new,num_data_new))
# 预测
print(pipe_model.predict(x_new).round(0))
```

## 7.4 使用Apriori关联分析提升商品销量

```python
# 导入库
from dataclasses import dataclass
import pandas as pd
import apriori
```

```python
# 定义关联类对象
@dataclass
class ASSO:
    sup: float = 0.01
    conf: float = 0.05

    @staticmethod
    def bulid_sets(data):
        # 将事务型数据转换为序列数据集
        order_data = data[['ORDER_ID', 'ITEM_ID']]
        order_list = order_data.groupby('ORDER_ID')['ITEM_ID'].unique().values.tolist()
        return [list(i) for i in order_list if len(list(i)) > 1]

    def generate_rule(self, order_list):
        L, suppData = apriori.apriori(order_list,minSupport=self.sup)
        return apriori.generateRules(order_list, L, suppData,minConf=self.conf)
```

```python
# 获取原始数据
raw_data = pd.read_excel('demo.xlsx',sheet_name=3)
print(raw_data.head(3))
```

```python
# 调用实例并输出关联结果
ass_model = ASSO(0.01,0.05)
order_list = ass_model.bulid_sets(raw_data)
rules = ass_model.generate_rule(order_list)
columns = ['ITEM_ID1', 'ITEM_ID2', 'INSTANCE', 'SUPPORT', 'CONFIDENCE','LIFT']
model_result = pd.DataFrame(rules, columns=columns)
print(model_result.head())
# 过滤lift>1的规则
valid_result = model_result[model_result['LIFT']>1]
print(valid_result.head())
```

```python
# 显示关联结果
from pyecharts.charts import Graph
from pyecharts import options as opts

# 找到所有的节点，即前项和后项商品集合并添加
node_data = valid_result['ITEM_ID1'].append(valid_result['ITEM_ID2'])
value_count = node_data.value_counts()
nodes = [opts.GraphNode(name=str(node), symbol_size=int(np.log(value)*10)) for node,value in zip(value_count.index,value_count.values)]
links = [opts.GraphLink(source=str(i), target=str(j)) for i,j in valid_result[['ITEM_ID1','ITEM_ID2']].values]
# 画出关系图
graph = Graph()
graph.add("", nodes, links, repulsion=2000)
graph.set_global_opts(title_opts=opts.TitleOpts(title="商品交叉销售关系"))
graph.render_notebook()
```

## 7.5 使用Prefixspan序列关联分析找到用户下一次最可能访问的页面

```python
from dataclasses import dataclass
import pandas as pd
from prefixspan import PrefixSpan as pfs
```

```python
@dataclass
class SEQV:
    mini_support: int = 2  # 最小支持度
    max_len: int = 2  # 序列最大项目数量

    @staticmethod
    def bulid_sets(data: pd.DataFrame):
        # 将事务型数据转换为序列数据集
        order_list = data.groupby('FULL_VISITOR_ID')['PAGE_URL'].unique().values.tolist()
        order_records = [list(i) for i in order_list]
        order_filter = [i for i in order_records if len(i) > 1]
        return order_filter

    def fit(self, frequent_seq_data):
        ps = pfs(frequent_seq_data)  # 建立对象
        # TODO 设置最长和最短序列长度--只要二项集，这是挖掘规则数量的关键
        ps.maxlen = ps.minlen = self.max_len
        rec_sequences = ps.frequent(self.mini_support)  # 获取最小支持度为2的频繁项集
        return rec_sequences

    def format_output(self, rec_sequences):
        target_sequence_items = [[i[0], i[1][0], i[1][1]] for i in
                                 rec_sequences]  # 格式化二项集
        sequences = pd.DataFrame(target_sequence_items, columns=['SUPPORT', 'ITEM_ID1','ITEM_ID2'])  # 建立数据框
        # 按支持度倒叙排序
        sequences_sort = sequences.sort_values(['SUPPORT'], ascending=False)
        return sequences_sort[['ITEM_ID1', 'ITEM_ID2', 'SUPPORT']]
```

```python
# 读取数据
page_data = pd.read_excel('demo.xlsx',sheet_name=4)
print(page_data.dtypes)
print(page_data.head())
```

```python
# 按FULL_VISITOR_ID 和VIEW_TIME排序
page_sort = page_data.sort_values(['FULL_VISITOR_ID','VIEW_TIME'])
print(page_sort.tail())
```

```python
# 应用序列关联分析
seq_model = SEQV(mini_support=2,max_len=2)
frequent_seq_data = seq_model.bulid_sets(page_sort)
rec_sequences = seq_model.fit(frequent_seq_data)
sequences_sort = seq_model.format_output(rec_sequences)
print(sequences_sort.head())
```

## 7.6 使用auto ARIMA时间序列预测线下门店销量

```python
# 导入库
import datetime
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller,arma_order_select_ic
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.tsa.arima_model import ARIMA
```

```python
# 读取数据
ts_data = pd.read_excel('demo.xlsx',sheet_name=5,index_col='DATE')['AMOUNT']
ts_data.head(3)
```

```python
# 平稳性检验
def adf_val(ts):
    return adfuller(ts)
    adf_se = pd.Series(adf_result, index=['statistic','p-value','lags','observations','critical values','icbest'])
    adf_se['critical values'] =adf_se['critical values'].round()
```

```python
print(adf_val(ts_data))
```

> 单位根检验adf值要分别小于critical_values中的1%、5%和10%的对应值且adf的p值<0.05

```python
# 白噪声检验
def stochastic_val(ts,lag):
    st_test = acorr_ljungbox(ts, lags=lag)
    return pd.Series(st_test,index=['lbvalue','pvalue'])
```

```python
print(stochastic_val(ts_data,1))
```

> 统计量的P值小于显著性水平0.05，则可以以95%的置信水平拒绝原假设，认为序列为非白噪声序列（否则，接受原假设，认为序列为纯随机序列。）

```python
# 获得最优PQ组合
def get_order(ts):
    return arma_order_select_ic(ts,max_ar=5,max_ma=5,ic=['aic', 'bic', 'hqic'])
```

```python
order = get_order(ts_data)
print(order.aic_min_order)
print(order.bic_min_order)
print(order.hqic_min_order)
```

```python
# 模型训练
def model_fit(ts,order):
    model = ARIMA(ts,order=order) #第二个参数代表使用了二阶差分
    results_ARIMA = model.fit()
    plt.figure(figsize=(10,4))
    plt.plot(ts)
    plt.plot(results_ARIMA.fittedvalues, 'r--') #红色线代表预测值
    plt.title('时间序列内训练结果。RMSE: %.2f'% np.sqrt(sum((results_ARIMA.fittedvalues-ts)**2)/ts.size))
    return results_ARIMA
```

```python
results_ARIMA = model_fit(ts_data,(5,0,2))
```

```python
# 预测
def predict(results_ARIMA,days):
    return results_ARIMA.forecast(days)[0]
```

```python
# 预测数据
days = 7
pre_value = predict(results_ARIMA,days)
index_value = [ts_data.index.max()+datetime.timedelta(i) for i in range(1,len(pre_value)+1)]
pre_data = pd.Series(pre_value,index=index_value)
print(pre_data)
# 显示图形
plt.figure(figsize=(10,4))
plt.plot(ts_data)
plt.plot(pre_data, 'r--') #红色线代表预测值
plt.title(f'未来{days}天销量图')
```

> 下面是封装为类的应用

```python
class AutoARIMA:
    # 平稳性检验
    @staticmethod
    def adf_val(ts):
        print('ADF validation:')
        dftest = adfuller(ts, autolag='AIC')
        dfoutput = pd.Series(dftest[0:4], index=['Statistic','p-value','#Lags Used','Number of Observations Used'])
        for key,value in dftest[4].items():
            dfoutput['Critical Value (%s)'%key] = value   #增加后面的显著性水平的临界值
        print(dfoutput)
        
    # 白噪声（随机性）检验
    @staticmethod
    def stochastic_val(ts,lag):
        print('stochastic validation:')
        print(acorr_ljungbox(ts, lags=lag)) #lags可自定义
        
    # 平稳性处理
    @staticmethod
    def diff(ts,n):
        return ts.diff(n)
    # 获得PQ值
    @staticmethod
    def get_order(ts):
        order = arma_order_select_ic(ts,max_ar=5,max_ma=5,ic=['aic', 'bic', 'hqic'])
        print(order.aic_min_order)
        print(order.bic_min_order)
        print(order.hqic_min_order)
    # arima
    @staticmethod
    def model_fit(ts,order):
        model = ARIMA(ts,order=order) #第二个参数代表使用了二阶差分
        results_ARIMA = model.fit(disp=-1)
        plt.figure(figsize=(10,4))
        plt.plot(ts)
        plt.plot(results_ARIMA.fittedvalues, 'r--') #红色线代表预测值
        plt.title('RMSE: %.2f'% np.sqrt(sum((results_ARIMA.fittedvalues-ts)**2)/ts.size))
        return results_ARIMA
    # 预测
    @staticmethod
    def predict(results_ARIMA,days):
        forecast_data = results_ARIMA.forecast(days)[0]
        #forecast_AR = results_ARIMA
        print(forecast_data)
```

## 7.7 使用IsolationForest异常检测找到异常广告流量

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.ensemble import IsolationForest
import pandas as pd
```

```python
# 读取数据
outlier_data = pd.read_excel('demo.xlsx',sheet_name=6,index_col='clientId')
print(outlier_data.shape)
```

```python
# 预处理
```

```python
# 去除全部为空的特征
data_dropna = outlier_data.dropna(axis='columns',how='all')
```

```python
# 找到NA列
cols_is_na = data_dropna.isnull().any(axis=0)
na_cols = cols_is_na[cols_is_na==True].index
print(data_dropna[na_cols].dtypes)
```

```python
# 填充NA列
print(data_dropna[na_cols].head())
fill_rules = {'newVisits': 0, 'pageviews': data_dropna['pageviews'].mean(), 'isVideoAd': 0, 'isTrueDirect': 0}
data_fillna = data_dropna.fillna(fill_rules)
print(data_fillna.isnull().any().sum())
```

```python
# 拆分数值型特征和字符串特征
str_or_num = data_fillna.dtypes=='object'
str_cols = str_or_num[str_or_num==True].index
string_data = data_fillna[str_cols]
num_data = data_fillna[[i for i in str_or_num.index if i not in str_cols]]
```

```python
# 分类特征转换为数值型索引
model_oe = OrdinalEncoder()
string_data_con = model_oe.fit_transform(string_data)
string_data_pd = pd.DataFrame(string_data_con,columns=string_data.columns,index=num_data.index)
```

```python
# 合并特征
feature_merge = pd.concat((num_data,string_data_pd),axis=1)
```

```python
# 模型检测
model_isof = IsolationForest(n_estimators=20, n_jobs=-1)
outlier_label = model_isof.fit_predict(feature_merge)
```

```python
# 输出异常数据值
outlier_data['outlier_label'] = outlier_label
outlier_count = outlier_data.groupby(['outlier_label'])['visitNumber'].count()
print(outlier_count)
```

```python
# 正常和异常结果在2个维度上的对比
import seaborn as sns
sns.jointplot(x="sessionQualityDim", y="hour", data=outlier_data[outlier_data['outlier_label']==1], kind="kde")
sns.jointplot(x="sessionQualityDim", y="hour", data=outlier_data[outlier_data['outlier_label']==-1], kind="kde")
```
