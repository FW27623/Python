```python
# 1）DataFrame至少10行，至少3列
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(10,3),index=list(range(0,10)),columns=list('ABC'))
print(df)
```

```python
# 2)DataFrame至少10行，至少3列
print(df.index)
print(df.columns)
```

```python
# 3）查看DataFrame的描述性统计结果
print(df.describe())
```

```python
# 4）按照索引排序（升序和降序）
#按照索引排序（升序）
print(df.sort_index())
#按照索引排序（降序）
print(df.sort_index(ascending=False))
```

```python
#5 )按照某一列进行排序（升序和降序）
#按照A列排序（升序）
print(df.sort_values(by='A'))
#按照A列排序（降序）
print(df.sort_values(by='A',ascending=False))
```

```python
#6) 使用列名查看A列的值
print(df.A)
```

```python
# 7）使用列名查看A列和B列的值
print(df.A,df.B)
```

```python
# 8）使用loc查看某个值
print(df.loc[0,'A'])
```

```python
# 9）使用loc查看2行和3列的值
#使用loc查看第2行的值
print(df.loc[1])
#使用loc查看第三列的值
print(df.loc[:,'C'])
```

```python
# 10）使用loc查看多行多列的值，如查看3行2列的值
print(df.loc[1,'C'])
```

```python
# 11）使用iloc查看某个值
print(df.iloc[0,0])
```

```python
# 12）使用iloc查看第三行的值和第二列的值
print(df.iloc[2])
print(df.iloc[:,1])
```

```python
# 13）使用iloc查看第3行第2列的值
print(df.iloc[2,1])
```

```python
# 14）为该DataFrame增加一个新列，且列值中存在小于0的数
df['D'] = df['A']*df['B']
print(df)
```

```python
# 15）获取新列中所有大于0的数
print(df[df['D']>0])
```

```python
# 16）计算新列的和
print(df['D'].sum())
```

```python
#16）计算新列的平均值
print(df['D'].mean())
```
